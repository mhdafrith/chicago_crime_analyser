```python
df=df.drop_duplicates(subset='Case Number', keep='first')
```

```python
df['Location Description'].fillna('OTHER', inplace=True)

df
```

```python
df['Location Description'] = df['Location Description'].replace('OTHER (SPECIFY)', 'OTHER')
```

```python
df=df.sort_values(by=['District', 'Ward', 'Community Area'], ascending=[True, True, True])
df['District'].fillna(method='ffill', inplace=True)
df['Ward'].fillna(method='ffill', inplace=True)
df['Community Area'].fillna(method='ffill', inplace=True)
df
```

```python
df=df.dropna()
```

```python
from google.colab import files

# Save DataFrame to a CSV file
df.to_csv('cleaned_chicago_crime_data.csv', index=False)

# Download the CSV file
files.download('cleaned_chicago_crime_data.csv')
```

```python
# Convert multiple columns from float to int
df[['Ward', 'Community Area']] = df[['Ward', 'Community Area']].astype('int')

print(df.dtypes)
```

```python
# Grouping the data by 'Primary Type' and counting the occurrences
bar_chart = df.groupby('Primary Type').size().sort_values(ascending=False).reset_index()

# Renaming columns for better readability
bar_chart.columns = ['Primary Type', 'Count']

# Displaying the top rows
bar_chart.head()
```

```python
import matplotlib.pyplot as plt

# Slicing the top 10 categories
top_10 = bar_chart.head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10['Primary Type'], top_10['Count'], color='skyblue')

# Adding data labels on top of the bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate (center of the bar)
        bar.get_height() + 5,              # Y-coordinate (slightly above the bar)
        str(int(bar.get_height())),        # Label (height of the bar)
        ha='center',                       # Horizontal alignment
        va='bottom',                       # Vertical alignment
        fontsize=10,                       # Font size
        color='black'                      # Text color
    )

# Formatting the chart
plt.xticks(rotation=90)
plt.title('Top 10 Crime Counts by Crime Type')
plt.xlabel('Primary Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
